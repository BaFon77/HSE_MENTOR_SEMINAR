```scala
%spark.pyspark 
from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```scala
%spark.pyspark
spark.sql("""
CREATE TABLE IF NOT EXISTS transactions_v2 (
    transaction_id   INT,
    user_id          INT,
    amount           DOUBLE,
    currency         STRING,
    transaction_date STRING,
    is_fraud         INT
)
STORED AS PARQUET
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS logs_v2 (
    log_id           INT,
    transaction_id   INT,
    category         STRING,
    comment          STRING,
    log_timestamp    STRING
)
STORED AS PARQUET
""")
```

```scala
%spark.pyspark
path = "s3a://hadoopbacket/files"

df_tx = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ",") \
    .csv(f"{path}/transactions_v2.csv")

df_tx.write.mode("overwrite").insertInto("transactions_v2")
print("transactions_v2:", df_tx.count(), "строк")

df_logs = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ";") \
    .csv(f"{path}/logs_v2.txt")

df_logs.write.mode("overwrite").insertInto("logs_v2")
print("logs_v2:", df_logs.count(), "строк")
```

```scala
%spark.pyspark
spark.sql("SELECT * FROM transactions_v2 LIMIT 5").show(truncate=False)
spark.sql("SELECT * FROM logs_v2 LIMIT 5").show(truncate=False)
```

```scala
%spark.pyspark
print("=== 1. Сумма транзакций по валютам USD, EUR, RUB ===")
spark.sql("""
    SELECT currency, COUNT(*) AS cnt, SUM(amount) AS total_amount
    FROM transactions_v2
    WHERE currency IN ('USD', 'EUR', 'RUB')
    GROUP BY currency
    ORDER BY total_amount DESC
""").show()

print("=== 2. Фрод-анализ ===")
spark.sql("""
    SELECT is_fraud,
           COUNT(*)        AS cnt,
           SUM(amount)     AS total_amount,
           AVG(amount)     AS avg_amount
    FROM transactions_v2
    GROUP BY is_fraud
""").show()

print("=== 3. Ежедневная статистика ===")
spark.sql("""
    SELECT transaction_date,
           COUNT(*)    AS daily_cnt,
           SUM(amount) AS daily_volume,
           AVG(amount) AS avg_amount
    FROM transactions_v2
    GROUP BY transaction_date
    ORDER BY transaction_date
""").show()

print("=== 4. Временные функции: день и месяц ===")
spark.sql("""
    SELECT transaction_id,
           transaction_date,
           MONTH(TO_DATE(transaction_date)) AS month,
           DAY(TO_DATE(transaction_date))   AS day,
           amount
    FROM transactions_v2
    ORDER BY month, day
""").show()

print("=== 5. JOIN с logs_v2 ===")
spark.sql("""
    SELECT t.transaction_id,
           t.amount,
           COUNT(l.log_id)   AS log_count,
           FIRST(l.category) AS top_category
    FROM transactions_v2 t
    LEFT JOIN logs_v2 l
        ON t.transaction_id = l.transaction_id
    GROUP BY t.transaction_id, t.amount
    ORDER BY log_count DESC
""").show()
```
