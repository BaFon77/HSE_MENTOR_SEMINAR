```scala
%spark.pyspark 
from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```scala
%spark.pyspark
users = spark.createDataFrame(
[ ("u1", "Berlin"),
("u2", "Berlin"),
("u3", "Munich"),
("u4", "Hamburg"), ],
["user_id", "city"] )
orders = spark.createDataFrame(
[ ("o1", "u1", "p1", 2, 10.0),
("o2", "u1", "p2", 1, 30.0),
("o3", "u2", "p1", 1, 10.0),
("o4", "u2", "p3", 5, 7.0),
("o5", "u3", "p2", 3, 30.0),
("o6", "u3", "p3", 1, 7.0),
("o7", "u4", "p1", 10, 10.0), ],
["order_id", "user_id", "product_id", "qty", "price"] )
products = spark.createDataFrame(
[ ("p1", "Ring VOLA"),
("p2", "Ring POROG"),
("p3", "Ring TISHINA"), ],
["product_id", "product_name"] )
```

```scala
%spark.pyspark
users.show() 
orders.show() 
products.show()
```

```scala
%spark.pyspark
orders_with_revenue = (
    orders
    .withColumn("revenue", F.col("qty") * F.col("price"))
)
```

```scala
%spark.pyspark
df = (
    orders_with_revenue
    .join(users, on="user_id", how="inner")
    .join(products, on="product_id", how="inner")
)
```

```scala
%spark.pyspark
mart_agg = (
    df
    .groupBy("city", "product_id", "product_name")
    .agg(
        F.count("order_id").alias("orders_cnt"),
        F.sum("qty").alias("qty_sum"),
        F.sum("revenue").alias("revenue_sum")
    )
)
```

```scala
%spark.pyspark
window_spec = Window.partitionBy("city").orderBy(F.col("revenue_sum").desc())

mart_city_top_products = (
    mart_agg
    .withColumn("rank", F.row_number().over(window_spec))
    .filter(F.col("rank") <= 2)
    .drop("rank")
)
```

```scala
%spark.pyspark
hadoop_conf = spark.sparkContext._jsc.hadoopConfiguration()
hadoop_conf.set("fs.s3a.endpoint", "storage.yandexcloud.net")
hadoop_conf.set("fs.s3a.signing-algorithm", "")
hadoop_conf.set("fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider")
hadoop_conf.set("fs.s3a.access.key","")
hadoop_conf.set("fs.s3a.secret.key","")
```

```scala
%spark.pyspark
path = "s3a://hadoopbacket/tmp/sandbox_zeppelin/mart_city_top_products"
mart_city_top_products.write.mode("overwrite").parquet(path)
```

```scala
%spark.pyspark
mart_result = spark.read.parquet(path)

mart_result.orderBy("city", F.col("revenue_sum").desc()).show(truncate=False)
```
